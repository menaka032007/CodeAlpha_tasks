```python
# Task 3: Data Visualization
# Tools: Python, Pandas, Matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# STEP 1: Define Data (No CSV file used)
data = {
    "Days": [1, 2, 3, 4, 5],
    "Sales": [100, 150, 130, 180, 200],
    "Category": ["A", "B", "C", "D"],
    "Values": [40, 30, 20, 10],
    "Study_Hours": [1, 2, 3, 4, 5],
    "Marks": [50, 55, 65, 70, 85]
}

df = pd.DataFrame({
    "Days": data["Days"],
    "Sales": data["Sales"],
    "Study_Hours": data["Study_Hours"],
    "Marks": data["Marks"]
})

# STEP 2: Line Chart
plt.figure()
plt.plot(df["Days"], df["Sales"])
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Line Chart - Sales Over Days")
plt.show()

# STEP 3: Bar Chart
plt.figure()
plt.bar(data["Category"], data["Values"])
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Bar Chart - Category Comparison")
plt.show()

# STEP 4: Pie Chart
plt.figure()
plt.pie(data["Values"], labels=data["Category"], autopct="%1.1f%%")
plt.title("Pie Chart - Category Distribution")
plt.show()

# STEP 5: Scatter Plot
plt.figure()
plt.scatter(df["Study_Hours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Scatter Plot - Study Hours vs Marks")
plt.show()

# STEP 6: Histogram
plt.figure()
plt.hist(df["Marks"])
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram - Marks Distribution")
plt.show()

print("Task 3 Data Visualization completed successfully")
```
