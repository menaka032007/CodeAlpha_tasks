```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- STEP 1: ASK MEANINGFUL QUESTIONS ----------------
print("STEP 1: Asking meaningful questions")
print("1. What is the average marks?")
print("2. Is there any relationship between study hours and marks?")
print("3. Are there any missing values?\n")

# ---------------- STEP 2: CREATE & EXPLORE DATA STRUCTURE ----------------
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Study_Hours': [2, 4, 6, 8, 10],
    'Marks': [50, 55, 65, 75, 90]
}

df = pd.DataFrame(data)

print("STEP 2: Data Structure")
print(df)
print("\nData Info:")
print(df.info())
print("\n")

# ---------------- STEP 3: IDENTIFY TRENDS & PATTERNS ----------------
print("STEP 3: Identifying trends and patterns")
print("Average Marks:", df['Marks'].mean())

plt.figure()
plt.plot(df['Study_Hours'], df['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Trend: Study Hours vs Marks")
plt.show()

# ---------------- STEP 4: TEST HYPOTHESIS ----------------
print("STEP 4: Hypothesis Testing")
print("Hypothesis: More study hours lead to higher marks")

correlation = df['Study_Hours'].corr(df['Marks'])
print("Correlation between Study Hours and Marks:", correlation)

plt.figure()
plt.scatter(df['Study_Hours'], df['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Hypothesis Validation")
plt.show()

# ---------------- STEP 5: DETECT DATA ISSUES ----------------
print("\nSTEP 5: Detecting data issues")
print("Missing values:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())

plt.figure()
plt.boxplot(df['Marks'])
plt.title("Outlier Detection - Marks")
plt.show()

print("\nTask 2: Exploratory Data Analysis completed successfully")
```
