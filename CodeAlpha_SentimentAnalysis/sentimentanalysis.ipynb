```python
# Task 4: Sentiment Analysis
# Project: Customer Feedback Analysis
# Tool: Python, TextBlob

from textblob import TextBlob

# STEP 1: Define project-related text data (Customer feedbacks)
feedbacks = [
    "The product quality is very good and I am satisfied",
    "Customer support response was slow and disappointing",
    "I love this application, it is very user friendly",
    "The service is bad and not worth the money",
    "Overall experience is okay, nothing special"
]

# STEP 2: Analyze sentiment for each feedback
print("Customer Feedback Sentiment Analysis\n")

for text in feedbacks:
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    # STEP 3: Classify sentiment
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    # STEP 4: Display output
    print("Feedback:", text)
    print("Polarity:", polarity)
    print("Sentiment:", sentiment)
    print("-" * 50)

# STEP 5: Conclusion
print("Task 4 completed: Sentiment analysis performed successfully")
```
